```python
# dropping the first column id
data=data.drop(labels='id',axis=1)

#x #y
x=data.drop(labels='diagnosis',axis=1)
y=data['diagnosis']
```

```python
#scalling convert all attribute to Z score

import scipy.stats as stats
xscaled=stats.zscore(x)
xscaled.describe()
```

```python

from sklearn.model_selection import train_test_split

#split the data
x_train,x_test,y_train,y_test=train_test_split(xscaled,y,test_size=0.30,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
# fit the model 
model.fit(x_train,y_train)
```

```python
#evaluate the performance 
predicted_labels=model.predict(x_test)
model.score(x_test,y_test)
```

```python
#calc accuracy  and confusion matrix
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,predicted_labels,labels=["M","B"])
plt.figure(figsize=(10,5))
df_cm=pd.DataFrame(cm)
sns.heatmap(df_cm,annot=True)
```
