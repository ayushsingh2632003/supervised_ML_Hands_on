```python

from scipy.stats import norm


for col in data.select_dtypes(include=[np.number]):
    
    md=data[col]
    
    mean=md.mean()
    std_dev=md.std()
    
    
    x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)
    y = norm.pdf(x, loc=mean, scale=std_dev)
    
    plt.plot(x, y)
    plt.xlabel(col)
    plt.ylabel('Probability Density')
    plt.title(f'Bell Curve for {col}')
    plt.show()

```

```python
# pedi anf test has outliers more therefore we will remove these outliers


data
```

```python
# Function to remove outliers using z-score
def remove_outliers(dataframe, column, z_thresh=3):
    z_scores = np.abs((dataframe[column] - dataframe[column].mean()) / dataframe[column].std())
    dataframe_cleaned = dataframe[z_scores < z_thresh]
    return dataframe_cleaned

# Specify the columns where you want to remove outliers
columns_to_remove_outliers = ['pedi','test']

# Apply the remove_outliers function to each specified column
for col in columns_to_remove_outliers:
    data = remove_outliers(data, col)

# Display the cleaned DataFrame without outliers
print(data)
```

```python
# outliers checking
plt.figure(figsize=(12, 8))
for i, column in enumerate(data.columns):
    plt.subplot(2, 4, i + 1)
    plt.boxplot(data[column])
    plt.title(column)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x=data.drop('class',axis=1)
x
```

```python
y=data['class']
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model=LogisticRegression(random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
model_score=model.score(x_test,y_test)
```

```python
print(model_score)
```

```python
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```
